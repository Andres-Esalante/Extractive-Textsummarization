```
import tensorflow.compat.v2 as tf
import tensorflow_datasets as tfds

# tfds works in both Eager and Graph modes
tf.enable_v2_behavior()

# Construct a tf.data.Dataset
cnn_dailymail_train = tfds.load('cnn_dailymail', split='train', shuffle_files=False)
cnn_dailymail_val   = tfds.load('cnn_dailymail', split='validation'  , shuffle_files=False)
cnn_dailymail_test  = tfds.load('cnn_dailymail', split='test' , shuffle_files=False) 
```

save the lenght of each split

```
cnn_dailymail_len_train = len(list(cnn_dailymail_train))
cnn_dailymail_len_val = len(list(cnn_dailymail_val))
cnn_dailymail_len_test = len(list(cnn_dailymail_test))
```

```
print(cnn_dailymail_len_train)
print(cnn_dailymail_len_val)
print(cnn_dailymail_len_test)
```

transform the data to list for each of the splits

```
import numpy as np

cnn_dailymail_train_data = []
cnn_dailymail_val_data = []
cnn_dailymail_test_data = []


for example in tfds.as_numpy(cnn_dailymail_train.take(cnn_dailymail_len_train)): 
  
    article = example["article"]
    article = article.decode("utf-8")

    highlights = example["highlights"]
    highlights = highlights.decode("utf-8")
    # make the highligh to be all in one line
    highlights = str.join(" ", highlights.splitlines())

    cnn_dailymail_train_data.append([highlights,article])


for example in tfds.as_numpy(cnn_dailymail_val.take(cnn_dailymail_len_val)): 
  
    article = example["article"]
    article = article.decode("utf-8")

    highlights = example["highlights"]
    highlights = highlights.decode("utf-8")
    highlights = str.join(" ", highlights.splitlines())

    cnn_dailymail_val_data.append([highlights,article])

for example in tfds.as_numpy(cnn_dailymail_test.take(cnn_dailymail_len_test)): 
  
    article = example["article"]
    article = article.decode("utf-8")

    highlights = example["highlights"]
    highlights = highlights.decode("utf-8")
    highlights = str.join(" ", highlights.splitlines())

    cnn_dailymail_test_data.append([highlights,article])


```

review length to check the sizes are correct

```
print(len(cnn_dailymail_train_data))
print(len(cnn_dailymail_val_data))
print(len(cnn_dailymail_test_data))
```

Check for empty news

```
# eliminate the indices with no news, (len <1)
list_del = []
for i in range(cnn_dailymail_len_train):
    if(len(cnn_dailymail_train_data[i][1])<10):
        list_del.append(i)
print(len(list_del))


list_del = []
for i in range(cnn_dailymail_len_val):
    if(len(cnn_dailymail_val_data[i][1])<10):
        list_del.append(i)
print(len(list_del))


list_del = []
for i in range(cnn_dailymail_len_test):
    if(len(cnn_dailymail_test_data[i][0])<10):
        list_del.append(i)
print(len(list_del))
```

see len in char of the data

```

avg_headline = 0
avg_news = 0

for i in range(len(cnn_dailymail_train_data)):
    avg_headline += len(cnn_dailymail_train_data[i][0])
    avg_news +=len(cnn_dailymail_train_data[i][1])

print("average len article: ",round(avg_news/len(cnn_dailymail_train_data)))
print("average len highlight: ",round(avg_headline/len(cnn_dailymail_train_data)))
```

```
import tensorflow.compat.v2 as tf
import tensorflow_datasets as tfds

# tfds works in both Eager and Graph modes
tf.enable_v2_behavior()

# Construct a tf.data.Dataset
multi_news_train = tfds.load('multi_news', split='train', shuffle_files=False)
multi_news_val   = tfds.load('multi_news', split='validation'  , shuffle_files=False)
multi_news_test  = tfds.load('multi_news', split='test' , shuffle_files=False) 
```

save the length

```
multi_news_len_train = len(list(multi_news_train))
multi_news_len_val = len(list(multi_news_val))
multi_news_len_test = len(list(multi_news_test))

print(multi_news_len_train)
print(multi_news_len_val)
print(multi_news_len_test)
```

transform the data to list for each of the splits

```
import numpy as np

multi_news_train_data = []
multi_news_val_data = []
multi_news_test_data = []


for example in tfds.as_numpy(multi_news_train.take(multi_news_len_train)): 
  
    article = example["document"]
    article = article.decode("utf-8")

    highlights = example["summary"]
    highlights = highlights.decode("utf-8")
    # make the highligh to be all in one line
    highlights = str.join(" ", highlights.splitlines())

    multi_news_train_data.append([highlights,article])


for example in tfds.as_numpy(multi_news_val.take(multi_news_len_val)): 
  
    article = example["document"]
    article = article.decode("utf-8")

    highlights = example["summary"]
    highlights = highlights.decode("utf-8")
    highlights = str.join(" ", highlights.splitlines())

    multi_news_val_data.append([highlights,article])

for example in tfds.as_numpy(multi_news_test.take(multi_news_len_test)): 
  
    article = example["document"]
    article = article.decode("utf-8")

    highlights = example["summary"]
    highlights = highlights.decode("utf-8")
    highlights = str.join(" ", highlights.splitlines())

    multi_news_test_data.append([highlights,article])


```

Review the length

```
print(len(multi_news_train_data))
print(len(multi_news_val_data))
print(len(multi_news_test_data))
```

check for empty news

```
# eliminate the indices with no news, (len <1)
list_del = []
for i in range(multi_news_len_train):
    if(len(multi_news_train_data[i][1])<10):
        list_del.append(i)
print(len(list_del))


list_del = []
for i in range(multi_news_len_val):
    if(len(multi_news_val_data[i][1])<10):
        list_del.append(i)
print(len(list_del))


list_del = []
for i in range(multi_news_len_test):
    if(len(multi_news_test_data[i][0])<10):
        list_del.append(i)
print(len(list_del))
```

remove |||||

```
for i in range(len(multi_news_train_data)):
    multi_news_train_data[i][1] = multi_news_train_data[i][1].replace("|||||", "")
    
for i in range(len(multi_news_val_data)):
    multi_news_val_data[i][1] = multi_news_val_data[i][1].replace("|||||", "")

for i in range(len(multi_news_test_data)):
    multi_news_test_data[i][1] = multi_news_test_data[i][1].replace("|||||", "")
```

```
avg_headline = 0
avg_news = 0

for i in range(len(multi_news_train_data)):
    avg_headline += len(multi_news_train_data[i][0])
    avg_news +=len(multi_news_train_data[i][1])

print("average len article: ",round(avg_news/len(multi_news_train_data)))
print("average len highlight: ",round(avg_headline/len(multi_news_train_data)))
```
