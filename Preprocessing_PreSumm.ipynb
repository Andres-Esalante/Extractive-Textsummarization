# Dataset 1

### 1.1 Convert to .story files

```python
import os
def create_story_files(df,num_file_start=0,num_file_end=2000,filename=None):
    for i in range(num_file_start,num_file_end):
        doc = df['text'][i] + '\n'*2 + '@highlight' + '\n'*2 + df['headlines'][i]
        file_name = os.path.join('',(str(i) + '.story'))
        print(file_name)
        with open(file_name,'w',encoding="ISO-8859-1") as story_file:
            story_file.write(doc)
```
